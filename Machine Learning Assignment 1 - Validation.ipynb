```python
# Initial Variables



# Input initial weight, separated by whitespace (theta1, theta2, theta3, theta4) 
# ex: 0.2 0.3 0.4 0.5

theta = [float(x) for x in input('Input initial weight = ').split()]
theta = np.array(theta)

                            
# Input initial bias
# ex: 0.9
bias = [float(x) for x in input('Input initial bias = ').split()]
bias = np.array(bias)

# Input alpha
alpha = float(input('Input initial alpha = '))
alpha

# Input number of epoch 
epoch = int(input('Input number of epoch = '))
epoch


error_each_epoch = np.zeros(60)
delta_theta_row = np.zeros(4)
```

```python
# Functions

def hFunc(theta,bias,r,data):
    return np.dot(data.iloc[r,:4],np.transpose(theta))+bias

def sigmoid(hFunc):
    return 1/(1+mt.exp(-hFunc))


def errorFunc(prediction,r,data):
    return (prediction-data.iloc[r,5])**2

def delta_theta(fact,prediction,x_row):
    return 2*(prediction-fact)*(1-prediction)*prediction*x_row

def new_theta(theta,alpha,delta_theta):
    return theta-alpha*delta_theta

def delta_bias(fact,prediction):
    return 2*(prediction-fact)*(1-prediction)*prediction

def new_bias(bias,alpha,delta_bias):
    return bias - alpha*delta_bias
```

```python
train_error_sum_1 = 0
valid_error_sum_1 = 0

train_error_sum_2 = 0
valid_error_sum_2 = 0


total_error_all_train_1 = np.zeros(60)
total_error_all_test_1 = np.zeros(60)


total_error_all_train_2 = np.zeros(60)
total_error_all_test_2 = np.zeros(60)

temp_theta_1 = np.zeros(4)
temp_bias_1 = 0
    
temp_theta_2 = np.zeros(4)
temp_bias_2 = 0
```

```python
for e in range(epoch):

    
    # Training
    for r in range(len(train_data_1)):
        x_row = np.array(train_data_1.iloc[r,:4]) 
        fact_row = np.array(train_data_1.iloc[r,5]) 
        
        
        prediction_val = hFunc(theta,bias,r,train_data_1)
        sigmoid_val = sigmoid(prediction_val)
        
        error_val = errorFunc(sigmoid_val,r,train_data_1)
        train_error_sum_1 += error_val

        
        
        # x = index for x1, x2, x3, x4
        for x in range(len(x_row)):
            delta_theta_row[x] = delta_theta(fact_row,sigmoid_val,x_row[x])
        
        # x = index for theta1, theta2, theta3, theta4
        for x in range(len(theta)):
            theta[x] = new_theta(theta[x],alpha,delta_theta_row[x])
        
        temp_theta_1 = theta




        delta_bias_val = delta_bias(fact_row,sigmoid_val)
        bias = new_bias(bias,alpha,delta_bias_val)
        temp_bias_1 = bias

    total_error_all_train_1[e] = train_error_sum_1
    train_error_sum_1 = 0
    
    # Validasi
    for r in range(len(test_data_1)):
        x_row = np.array(test_data_1.iloc[r,:4]) 
        fact_row = np.array(test_data_1.iloc[r,5]) 
        
        
        prediction_val = hFunc(temp_theta_1,temp_bias_1,r,train_data_1)
        sigmoid_val = sigmoid(prediction_val)
        
        error_val = errorFunc(sigmoid_val,r,test_data_1)
        valid_error_sum_1 += error_val

        

    total_error_all_test_1[e] = valid_error_sum_1
    valid_error_sum_1 = 0
    
    
    for r in range(len(train_data_1)):
        x_row = np.array(train_data_1.iloc[r,:4]) 
        fact_row = np.array(train_data_1.iloc[r,5]) 
        
        
        prediction_val = hFunc(theta,bias,r,train_data_1)
        sigmoid_val = sigmoid(prediction_val)
        
        error_val = errorFunc(sigmoid_val,r,train_data_1)
        train_error_sum_2 += error_val

        
        
        # x = index for x1, x2, x3, x4
        for x in range(len(x_row)):
            delta_theta_row[x] = delta_theta(fact_row,sigmoid_val,x_row[x])
        
        # x = index for theta1, theta2, theta3, theta4
        for x in range(len(theta)):
            theta[x] = new_theta(theta[x],alpha,delta_theta_row[x])
        
        temp_theta_2 = theta




        delta_bias_val = delta_bias(fact_row,sigmoid_val)
        bias = new_bias(bias,alpha,delta_bias_val)
        temp_bias_2 = bias

    total_error_all_train_2[e] = train_error_sum_2
    train_error_sum_2 = 0
    
    
    # Validasi
    for r in range(len(test_data_1)):
        x_row = np.array(test_data_1.iloc[r,:4]) 
        fact_row = np.array(test_data_1.iloc[r,5]) 
        
        
        prediction_val = hFunc(temp_theta_2,temp_bias_2,r,train_data_1)
        sigmoid_val = sigmoid(prediction_val)
        
        error_val = errorFunc(sigmoid_val,r,test_data_1)
        valid_error_sum_2 += error_val

        

    total_error_all_test_2[e] = valid_error_sum_2
    valid_error_sum_2 = 0
    
```

```python
total_error_banget_train = np.add(total_error_all_train_1,total_error_all_train_2)
total_error_banget_train_tambah = total_error_all_train_1 + total_error_all_train_2
total_error_banget_test = np.add(total_error_all_test_1,total_error_all_test_2)
# print(total_error_all_train_1)
```
