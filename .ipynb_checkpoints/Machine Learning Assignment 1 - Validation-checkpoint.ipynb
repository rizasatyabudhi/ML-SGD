```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math as mt

        
dataset = pd.read_csv('irisdata.csv')
dataset
```

```python
# Add new class column

def label(row):
    if row['label'] == 'Iris-setosa':
        return 0
    else:
        return 1

dataset['class'] = dataset.apply (lambda row: label(row),axis=1)
dataset 
# dataset_np = np.array(dataset)


```

```python
from sklearn.model_selection import KFold
data = pd.read_csv('irisdata.csv')

kf = KFold(n_splits=5,shuffle=True,random_state=np.random)

for train, test in kf.split(dataset):
    train_data_1 = np.array(dataset)[train]
    test_data_1 = np.array(dataset)[test]
    

for train, test in kf.split(dataset):
    train_data_2 = np.array(dataset)[train]
    test_data_2 = np.array(dataset)[test]
    
    
for train, test in kf.split(dataset):
    train_data_3 = np.array(dataset)[train]
    test_data_3 = np.array(dataset)[test]
    
    
for train, test in kf.split(dataset):
    train_data_4 = np.array(dataset)[train]
    test_data_4 = np.array(dataset)[test]
    
for train, test in kf.split(dataset):
    train_data_5 = np.array(dataset)[train]
    test_data_5 = np.array(dataset)[test]
    


train_data_1 = pd.DataFrame(train_data_1)
test_data_1 = pd.DataFrame(test_data_1)

train_data_2 = pd.DataFrame(train_data_2)
test_data_2 = pd.DataFrame(test_data_2)


train_data_3 = pd.DataFrame(train_data_3)
test_data_3 = pd.DataFrame(test_data_3)

train_data_4 = pd.DataFrame(train_data_4)
test_data_4 = pd.DataFrame(test_data_4)

train_data_5 = pd.DataFrame(train_data_5)
test_data_5 = pd.DataFrame(test_data_5)
```

```python
# Initial Variables


# Input initial weight, separated by whitespace (theta1, theta2, theta3, theta4) 
# ex: 0.2 0.3 0.4 0.5

theta = [float(x) for x in input('Input initial weight = ').split()]
theta = np.array(theta)

                            
# Input initial bias
# ex: 0.9
bias = [float(x) for x in input('Input initial bias = ').split()]
bias = np.array(bias)

# Input alpha
alpha = float(input('Input initial alpha = '))
alpha

# Input number of epoch 
epoch = int(input('Input number of epoch = '))
epoch


error_each_epoch = np.zeros(60)
delta_theta_row = np.zeros(4)
```

```python
# Functions

def hFunc(theta,bias,r,data):
    return np.dot(data.iloc[r,:4],np.transpose(theta))+bias

def sigmoid(hFunc):
    return 1/(1+mt.exp(-hFunc))

def errorFunc(prediction,r,data):
    return (prediction-data.iloc[r,5])**2

def delta_theta(fact,prediction,x_row):
    return 2*(prediction-fact)*(1-prediction)*prediction*x_row

def new_theta(theta,alpha,delta_theta):
    return theta-alpha*delta_theta

def delta_bias(fact,prediction):
    return 2*(prediction-fact)*(1-prediction)*prediction

def new_bias(bias,alpha,delta_bias):
    return bias - alpha*delta_bias
```

```python
train_error_sum_1 = 0
valid_error_sum_1 = 0

train_error_sum_2 = 0
valid_error_sum_2 = 0

train_error_sum_3 = 0
valid_error_sum_3 = 0

train_error_sum_4 = 0
valid_error_sum_4 = 0

train_error_sum_5 = 0
valid_error_sum_5 = 0



total_error_all_train_1 = np.zeros(60)
total_error_all_test_1 = np.zeros(60)


total_error_all_train_2 = np.zeros(60)
total_error_all_test_2 = np.zeros(60)


total_error_all_train_3 = np.zeros(60)
total_error_all_test_3 = np.zeros(60)

total_error_all_train_4 = np.zeros(60)
total_error_all_test_4 = np.zeros(60)

total_error_all_train_5 = np.zeros(60)
total_error_all_test_5 = np.zeros(60)


temp_theta_1 = np.zeros(4)
temp_bias_1 = 0
    
temp_theta_2 = np.zeros(4)
temp_bias_2 = 0

temp_theta_3 = np.zeros(4)
temp_bias_3 = 0

temp_theta_4 = np.zeros(4)
temp_bias_4 = 0

temp_theta_5 = np.zeros(4)
temp_bias_5 = 0
```

```python
for e in range(epoch):

    
    # Training Fold 1
    for r in range(len(train_data_1)):
        x_row = np.array(train_data_1.iloc[r,:4]) 
        fact_row = np.array(train_data_1.iloc[r,5]) 
        
        
        prediction_val = hFunc(theta,bias,r,train_data_1)
        sigmoid_val = sigmoid(prediction_val)
        
        error_val = errorFunc(sigmoid_val,r,train_data_1)
        train_error_sum_1 += error_val

        
        
        # x = index for x1, x2, x3, x4
        for x in range(len(x_row)):
            delta_theta_row[x] = delta_theta(fact_row,sigmoid_val,x_row[x])
        
        # x = index for theta1, theta2, theta3, theta4
        for x in range(len(theta)):
            theta[x] = new_theta(theta[x],alpha,delta_theta_row[x])
        
        temp_theta_1 = theta


        delta_bias_val = delta_bias(fact_row,sigmoid_val)
        bias = new_bias(bias,alpha,delta_bias_val)
        temp_bias_1 = bias

    total_error_all_train_1[e] = train_error_sum_1 
    train_error_sum_1 = 0
    
    # Validation Fold 1
    for r in range(len(test_data_1)):
        x_row = np.array(test_data_1.iloc[r,:4]) 
        fact_row = np.array(test_data_1.iloc[r,5]) 
        
        
        prediction_val = hFunc(temp_theta_1,temp_bias_1,r,test_data_1)
        sigmoid_val = sigmoid(prediction_val)
        
        error_val = errorFunc(sigmoid_val,r,test_data_1)
        valid_error_sum_1 += error_val

        

    total_error_all_test_1[e] = valid_error_sum_1 
    valid_error_sum_1 = 0
    
    
    # Training Fold 2
    for r in range(len(train_data_2)):
        x_row = np.array(train_data_1.iloc[r,:4]) 
        fact_row = np.array(train_data_1.iloc[r,5]) 
        
        
        prediction_val = hFunc(theta,bias,r,train_data_2)
        sigmoid_val = sigmoid(prediction_val)
        
        error_val = errorFunc(sigmoid_val,r,train_data_2)
        train_error_sum_2 += error_val

        
        
        # x = index for x1, x2, x3, x4
        for x in range(len(x_row)):
            delta_theta_row[x] = delta_theta(fact_row,sigmoid_val,x_row[x])
        
        # x = index for theta1, theta2, theta3, theta4
        for x in range(len(theta)):
            theta[x] = new_theta(theta[x],alpha,delta_theta_row[x])
        
        temp_theta_2 = theta




        delta_bias_val = delta_bias(fact_row,sigmoid_val)
        bias = new_bias(bias,alpha,delta_bias_val)
        temp_bias_2 = bias

    total_error_all_train_2[e] = train_error_sum_2 
    train_error_sum_2 = 0
    
    
    # Validation Fold 2
    for r in range(len(test_data_2)):
        x_row = np.array(test_data_2.iloc[r,:4]) 
        fact_row = np.array(test_data_2.iloc[r,5]) 
        
        
        prediction_val = hFunc(temp_theta_2,temp_bias_2,r,test_data_2)
        sigmoid_val = sigmoid(prediction_val)
        
        error_val = errorFunc(sigmoid_val,r,test_data_2)
        valid_error_sum_2 += error_val

        

    total_error_all_test_2[e] = valid_error_sum_2 
    valid_error_sum_2 = 0
    
     
    
   # Training Fold 3
    for r in range(len(train_data_1)):
        x_row = np.array(train_data_3.iloc[r,:4]) 
        fact_row = np.array(train_data_3.iloc[r,5]) 
        
        
        prediction_val = hFunc(theta,bias,r,train_data_3)
        sigmoid_val = sigmoid(prediction_val)
        
        error_val = errorFunc(sigmoid_val,r,train_data_3)
        train_error_sum_3 += error_val

        
        
        # x = index for x1, x2, x3, x4
        for x in range(len(x_row)):
            delta_theta_row[x] = delta_theta(fact_row,sigmoid_val,x_row[x])
        
        # x = index for theta1, theta2, theta3, theta4
        for x in range(len(theta)):
            theta[x] = new_theta(theta[x],alpha,delta_theta_row[x])
        
        temp_theta_3 = theta




        delta_bias_val = delta_bias(fact_row,sigmoid_val)
        bias = new_bias(bias,alpha,delta_bias_val)
        temp_bias_3 = bias

    total_error_all_train_3[e] = train_error_sum_3 
    train_error_sum_3 = 0
    
    
    # Validation Fold 3
    for r in range(len(test_data_3)):
        x_row = np.array(test_data_3.iloc[r,:4]) 
        fact_row = np.array(test_data_3.iloc[r,5]) 
        
        
        prediction_val = hFunc(temp_theta_3,temp_bias_3,r,train_data_3)
        sigmoid_val = sigmoid(prediction_val)
        
        error_val = errorFunc(sigmoid_val,r,test_data_3)
        valid_error_sum_3 += error_val

        

    total_error_all_test_3[e] = valid_error_sum_3 
    valid_error_sum_3 = 0
    
    
# Training Fold 4
    for r in range(len(train_data_4)):
        x_row = np.array(train_data_4.iloc[r,:4]) 
        fact_row = np.array(train_data_4.iloc[r,5]) 
        
        
        prediction_val = hFunc(theta,bias,r,train_data_4)
        sigmoid_val = sigmoid(prediction_val)
        
        error_val = errorFunc(sigmoid_val,r,train_data_4)
        train_error_sum_4 += error_val

        
        
        # x = index for x1, x2, x3, x4
        for x in range(len(x_row)):
            delta_theta_row[x] = delta_theta(fact_row,sigmoid_val,x_row[x])
        
        # x = index for theta1, theta2, theta3, theta4
        for x in range(len(theta)):
            theta[x] = new_theta(theta[x],alpha,delta_theta_row[x])
        
        temp_theta_4 = theta




        delta_bias_val = delta_bias(fact_row,sigmoid_val)
        bias = new_bias(bias,alpha,delta_bias_val)
        temp_bias_4 = bias

    total_error_all_train_4[e] = train_error_sum_4 
    train_error_sum_4 = 0
    
    
    # Validation Fold 4
    for r in range(len(test_data_4)):
        x_row = np.array(test_data_4.iloc[r,:4]) 
        fact_row = np.array(test_data_4.iloc[r,5]) 
        
        
        prediction_val = hFunc(temp_theta_4,temp_bias_4,r,train_data_4)
        sigmoid_val = sigmoid(prediction_val)
        
        error_val = errorFunc(sigmoid_val,r,test_data_4)
        valid_error_sum_4 += error_val

        

    total_error_all_test_4[e] = valid_error_sum_4 
    valid_error_sum_4 = 0
    

# Training Fold 5
    for r in range(len(train_data_4)):
        x_row = np.array(train_data_5.iloc[r,:4]) 
        fact_row = np.array(train_data_5.iloc[r,5]) 
        
        
        prediction_val = hFunc(theta,bias,r,train_data_5)
        sigmoid_val = sigmoid(prediction_val)
        
        error_val = errorFunc(sigmoid_val,r,train_data_5)
        train_error_sum_5 += error_val

        
        
        # x = index for x1, x2, x3, x4
        for x in range(len(x_row)):
            delta_theta_row[x] = delta_theta(fact_row,sigmoid_val,x_row[x])
        
        # x = index for theta1, theta2, theta3, theta4
        for x in range(len(theta)):
            theta[x] = new_theta(theta[x],alpha,delta_theta_row[x])
        
        temp_theta_5 = theta




        delta_bias_val = delta_bias(fact_row,sigmoid_val)
        bias = new_bias(bias,alpha,delta_bias_val)
        temp_bias_5 = bias

    total_error_all_train_5[e] = train_error_sum_5
    train_error_sum_5 =  0
    
    
    # Validation Fold 5
    for r in range(len(test_data_5)):
        x_row = np.array(test_data_5.iloc[r,:4]) 
        fact_row = np.array(test_data_5.iloc[r,5]) 
        
        
        prediction_val = hFunc(temp_theta_5,temp_bias_5,r,train_data_5)
        sigmoid_val = sigmoid(prediction_val)
        
        error_val = errorFunc(sigmoid_val,r,test_data_5)
        valid_error_sum_5 += error_val

        

    total_error_all_test_5[e] = valid_error_sum_5 
    valid_error_sum_5 = 0
```

```python
total_error_final_train = (total_error_all_train_1 + total_error_all_train_2+total_error_all_train_3 + total_error_all_train_4 + total_error_all_train_5)
total_error_final_test = (total_error_all_test_1 + total_error_all_test_2 + total_error_all_test_3 + total_error_all_test_4 + total_error_all_test_5)
```

```python
# Visualization

import matplotlib.pyplot as plt

plt.plot(total_error_final_train, color="blue")
plt.plot(total_error_final_test, color="green")
plt.title("Error graph after using 5 fold (0.8)")
plt.legend(["Data Training", "Data Validation"])
plt.ylabel('error')
plt.xlabel('epoch')
```
